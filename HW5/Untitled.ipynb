```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

cm=np.array([[1,2,3],[7,8,9],[4,5,6]])
df=pd.DataFrame(cm,index=['a','b','c'],columns=['d','e','f'])
plt.figure(figsize=(5,5))
sns.heatmap(df,annot=True,cmap="YlGnBu")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Example plot")
plt.show()
```
